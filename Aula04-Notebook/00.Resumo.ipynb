## LISTA

```python
# Lista - Pode ser modificado a qualquer momento
# Lista usa [] para defenir valores
# possível criar uma variável com vários valores

nome = ["Felipe","Claudio","Mário"]
# cada valor tem sua posição definida - conhecido como indíce que começa com ZERO
# Felipe está no indice 0 - Claudio 1 - Mario 2
print(nome[0])

# .append("") # é utilizado para adicionar novo valor
nome.append("João")
print(nome)

# para motificar algum valor existente:
nome[1] = "Jose"
print(nome)
```

## TUPLA

```python
# Tupla é uma lista que não pode ser alterado depois de criado.
# Tupla usa () para defenir valores

cores = ("Azul","Verde","Vermelho")
print(cores)

# para selecionar um valor específico
print(cores[1]) # Lembrando que inicia em ZERO
```

## DICIONÁRIO

```python
# Dicionário usa {}
# Deixa de usar indíce (igual Lista e Tupla) passar a usar chave

aluno = {
    "nome": "Professor", "idade": 20, "cidade": "Jaraguá do Sul"
}

print(aluno["nome"])

# Pode incluir valores
aluno["curso"] = "Análise de Dados com Python"
print(aluno)

# Pode alterar valores:
aluno["idade"] = 39
print(aluno)
```

```python
# Exemplo 1 — Lista de compras
# Vamos criar uma lista, acessar o primeiro e o último item pelos índices, e 
# testar os 5 métodos da tabela na prática — vendo o que cada um faz e em qual situação você usaria ele:
compras = ["arroz","feijão","leite"]
print(compras[0])
print(compras[-1])

# curiosidade max e mim na lista de str (string) pega por ordem alfabética
print(max(compras))
print(min(compras))

# .append() adicionar no final da lista
compras.append("café")
print(compras)

# .insert() adiciona na posição informada
compras.insert(0,"manteiga")
print(compras)

# .remove() remover o valor desejado
compras.remove("café")
print(compras)

# .pop() retirar da lista, porém mostrando o valor
item_retirado = compras.pop()
print(f"Retirado da lista: {item_retirado}")
print(compras)

# .sort() reorganiza a lista em ordem alfabética
compras.sort()
print(compras)
```

```python
# Exemplo 2 — Calculando a média de uma lista de notas
# Agora veja como usar len() para contar os itens e sum() para somar todos os valores
# combinando os dois, calculamos a média da turma em uma linha:

notas = [7.5, 8.0, 6.5, 9.0, 5.5]
media = sum(notas)/len(notas)
print(f"Média da turma é: {media:.2}")

```
