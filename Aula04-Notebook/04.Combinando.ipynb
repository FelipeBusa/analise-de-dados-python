## 4. Combinando tudo: o carrinho de compras

Um carrinho de compras é uma **lista de dicionários** — cada item tem produto, preço e quantidade.

### ✏️ Atividade Prática 4 — Sua vez de programar

Crie o seu próprio carrinho de compras (lista de dicionários) com pelo menos 3 itens e calcule o total.

```python
compras = [
    {"item": "Feijão", "preco": 10.0, "quantidade": 1},
    {"item": "Arroz", "preco": 5.0, "quantidade": 2},
    {"item": "Carne", "preco": 28.00,"quantidade": 3}
]
total = 0
for produto in compras: 
    total += produto["preco"] * produto["quantidade"]
print(f"O valor total da compra é: R$ {total:.2f}")

# Exemplo do total por produto
# for produto in compras:
#     subtotal = produto["preco"] * produto["quantidade"]
#     print(f"O valor por produto é: {produto["item"]}, {produto["quantidade"]} quantidade. Total: R$ {subtotal:.2f}")   
```
