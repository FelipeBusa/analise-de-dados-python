```python
# ✏️ Atividade 1 - Contar por setor (CSV)
# Usando `cadastro_funcionario.csv`, filtre e conte quantos funcionários existem no setor `"Estoque"`.
# Replique para outras colunas e informações

with open("cadastro_funcionario.csv", encoding="utf-8") as arquivo:
    leitor = csv.DictReader(arquivo)
    funcionarios =  list(leitor)

contador_setor = 0
contador_cargo = 0

for funcionario in funcionarios:
    if funcionario['setor'] == "Estoque":
        contador_setor += 1

    if funcionario['cargo'] == "Atendente":
        contador_cargo += 1

print(f"Total de funcionários no setor 'Estoque': {contador_setor}")
print(f"Total de funcionários no cargo 'Atendente': {contador_cargo}")
```

### ✏️ Atividade 2 — Escrever um resumo (JSON)

A partir do CSV, monte um dicionário contando quantos funcionários existem em cada setor e salve esse resumo em um novo arquivo `relatorio_setores.json`.

```python
# Atividade pelo professor
with open("cadastro_funcionario.csv", encoding="utf-8") as arquivo:
    leitor = csv.DictReader(arquivo)
    funcionarios =  list(leitor)
    
contagem_setores = {}

for f in funcionarios:
    setor = f["setor"]
    contagem_setores[setor] = contagem_setores.get(setor,0) + 1

resumo = {"total_funcionarios": len(funcionarios), "por setor":
    contagem_setores}

with open("relatorio_setoresv1.json", "w", encoding="utf-8") as arquivo:
    json.dump(resumo,arquivo, ensure_ascii=False, indent=4)
    
print(resumo)
```

```python
with open("cadastro_funcionario.csv", encoding="utf-8") as arquivo:
    leitor = csv.DictReader(arquivo)
    funcionarios =  list(leitor)


setores = {}

for funcionario in funcionarios:
    setor = funcionario["setor"]
    if setor not in setores:
        setores[setor] = 0
    setores[setor] += 1

print("Setores:")
for setor in setores:
    print(f"{setor}: {setores[setor]} funcionário(s)")

with open("relato_setores.json", "w", encoding="utf-8") as arquivo_json:
    json.dump(setores, arquivo_json, ensure_ascii=False, indent=4)
```

## Exemplo 1 - Data hora


```python
from datetime import datetime

# registro real de Ana Souza (id 1) em 02/01/2023
entrada = datetime(2023, 1, 2, 8, 30)
saida = datetime(2023, 1, 2, 16, 25)

horas_trabalhadas = (saida - entrada).seconds / 3600
print(f"Horas trabalhadas: {horas_trabalhadas:.2f}h")
```

## Exemplo 2 - Calculando o atraso em minutos

```python
from datetime import datetime

entrada_prevista_min = 8 * 60  # 08:00, em minutos
entrada_real = datetime(2023, 1, 2, 8, 30)
entrada_real_min = entrada_real.hour * 60 + entrada_real.minute

atraso_minutos = entrada_real_min - entrada_prevista_min
print(f"Atraso: {atraso_minutos} minutos")
```
