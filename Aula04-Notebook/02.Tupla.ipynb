## 2. Tuplas

Uma **tupla** é parecida com uma lista, mas usa **parênteses** `( )` em vez de colchetes — e a diferença fundamental é que ela é **imutável**: depois de criada, os valores não podem ser alterados.

```python
cores  = ("vermelho", "verde", "azul")   # tupla de strings
ponto  = (10, 20)                         # coordenada x, y
data   = (15, 3, 2005)                    # dia, mês, ano
```

**Quando usar tupla em vez de lista?**  
Use tupla para dados que *não devem mudar nunca*: coordenadas geográficas, data de nascimento, configurações fixas de um sistema. A imutabilidade **protege o dado** de alterações acidentais — o Python recusa qualquer tentativa de mudança e lança um erro.

> 💡 Tupla acessa itens pelos mesmos índices da lista: `ponto[0]` → `10`, `ponto[-1]` → `20`.

### 🔹 Exemplo 1 — Dados que não devem mudar

Vamos criar uma tupla de coordenadas, acessar seus valores e, depois, tentar alterar um item para ver o erro que o Python lança:

````python
coordenadas = (10,20)
print(f"X = {coordenadas[0]}, Y = {coordenadas[1]}")
````

### 🔹 Exemplo 2 — Tupla para data de nascimento

Uma data de nascimento nunca muda — caso perfeito para tupla. Veja como criar a tupla e extrair cada parte em variáveis separadas para montar uma frase com f-string:

````python
nascimento = (16,7,2026)
dia = nascimento[0]
mes = nascimento[1]
ano = nascimento[2]

print(f"Data de nascimento é: {dia}/{mes}/{ano}")
````
